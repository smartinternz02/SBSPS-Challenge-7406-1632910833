IMPORTING DATA 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data=pd.read_csv("Sanitation.csv")
label=[]
for row in data.ULB:
    label.append(row)
label
data
```

PIE CHART

```python
textprops={"fontsize":15}
fig=plt.pie(data.Garbagelifted,labels=label,radius=20,textprops=textprops,autopct = "%0.2f%%")
plt.title('waste')
plt.legend()
plt.show(fig)
```

```python
LINE PLOT
```

```python
fig1=plt.plot(data.Garbagegeneration,label,color='r')
fig1=plt.plot(data.Garbagelifted,label,color='g')
plt.rc('lines', linewidth=2, linestyle='-', marker='*')
plt.rcParams["figure.figsize"] = (50, 30)
plt.xlabel('Garbage Generation')
plt.ylabel('Garbage Lifted')
plt.grid(fig1)
```

```python
AREA PLOT
```

```python
data.plot.area(x='ULB')
```

```python
df=pd.DataFrame(columns=['Difference'])
df['Difference']=data['Garbagegeneration']-data['Garbagelifted']
df
```
